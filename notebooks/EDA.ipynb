```python
import pandas as pd
import numpy as np
import os, sys

os.getcwd()
rpath = os.path.abspath('../')
if rpath not in sys.path:
    sys.path.insert(0,rpath)

# Load the .txt file
# Construct the file path
file_path = os.path.join(rpath, 'data', 'MachineLearningRating_v3.txt')

# Skip lines with too many fields
# Try with different delimiters
data = pd.read_csv(file_path, delimiter='|')  # Tab-delimited
data.head()
```

```python
def handeling_missing_data(data):
    data.replace([np.inf, -np.inf], np.nan, inplace=True)
    missing_data = data.isnull().sum().sort_values(ascending=False)
    missing_percentage = (missing_data / len(data)) * 100
    data_cleaned = data.drop(columns=missing_data[missing_percentage > 90].index)

    # Step 3: Impute missing values in numeric columns using the mean
    numeric_cols = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
    data_cleaned[numeric_cols] = data_cleaned[numeric_cols].fillna(0)

    # Step 4: Impute missing values in categorical columns using 'unknown'
    categorical_cols = data_cleaned.select_dtypes(include=['object']).columns
    data_cleaned[categorical_cols] = data_cleaned[categorical_cols].fillna('unknown')
    return data_cleaned
```

```python
data_cleaned = handeling_missing_data(data)
data_cleaned.isnull().sum()
```

```python
# Descriptive statistics for numerical columns
numerical_cols = ['TotalClaims', 'TotalPremium', 'SumInsured', 'CalculatedPremiumPerTerm']
data_cleaned[numerical_cols].describe()
```

```python
# Summary of categorical variables
categorical_cols = ['Citizenship', 'VehicleType', 'CoverType', 'Gender']
for col in categorical_cols:
    print(f"{col} value counts:\n{data_cleaned[col].value_counts()}\n")
```

```python
# Convert to categorical type
data_cleaned['Citizenship'] = data_cleaned['Citizenship'].astype('category')
data_cleaned['VehicleType'] = data_cleaned['VehicleType'].astype('category')
```

```python
# Check for duplicates
data_cleaned.duplicated().sum()

# Remove duplicates
data_cleaned = data_cleaned.drop_duplicates()
```

```python
data_cleaned['ClaimToPremiumRatio'] = data_cleaned['TotalClaims'] / data_cleaned['TotalPremium']
data_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical and categorical columns
numerical_columns = data_cleaned.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = data_cleaned.select_dtypes(include=['object', 'bool']).columns

# Plot histograms for numerical columns
def plot_histograms(df, columns):
    df[columns].hist(figsize=(15, 10), bins=30, color='skyblue', edgecolor='black')
    plt.suptitle('Histograms of Numerical Features', size=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# Plot bar charts for categorical columns
def plot_bar_charts(df, columns):
    fig, axes = plt.subplots(len(columns), 1, figsize=(10, len(columns)*5))
    for i, column in enumerate(columns):
        if df[column].nunique() < 20:  # Plot bar charts only for columns with fewer than 20 unique values
            sns.countplot(x=column, data_cleaned=df, ax=axes[i], palette='Set2')
            axes[i].set_title(f'Distribution of {column}', size=14)
            axes[i].set_ylabel('Count')
            axes[i].set_xlabel(column)
            axes[i].tick_params(axis='x', rotation=45)
    plt.tight_layout()
    plt.show()

# Call the functions
plot_histograms(data_cleaned, numerical_columns)
plot_bar_charts(data_cleaned, categorical_columns)
```

```python
# Ensure the required columns are numerical
data_cleaned[['TotalPremium', 'TotalClaims']] = data_cleaned[['TotalPremium', 'TotalClaims']].apply(pd.to_numeric, errors='coerce')

# Drop rows with missing or NaN values for TotalPremium, TotalClaims, and PostalCode
data_cleaned = data_cleaned.dropna(subset=['TotalPremium', 'TotalClaims', 'PostalCode'])

# Convert PostalCode to a categorical variable if needed
data_cleaned['PostalCode'] = data_cleaned['PostalCode'].astype(str)

# Scatter plot: TotalPremium vs. TotalClaims as a function of PostalCode
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_cleaned, x='TotalPremium', y='TotalClaims', hue='PostalCode', palette='viridis', s=100, alpha=0.7)
plt.title('Scatter Plot of TotalPremium vs. TotalClaims by PostalCode', size=16)
plt.xlabel('Total Premium', size=14)
plt.ylabel('Total Claims', size=14)
plt.legend(title='Postal Code', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Calculate the correlation matrix
correlation_matrix = data_cleaned[['TotalPremium', 'TotalClaims']].corr()

# Plot correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix: TotalPremium and TotalClaims', size=16)
plt.show()
```

```python

# Plot 1: Compare CoverType distribution over PostalCode
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='PostalCode', hue='CoverType', palette='Set1')
plt.title('Distribution of CoverType Across Postal Codes', size=16)
plt.xlabel('Postal Code', size=14)
plt.ylabel('Count', size=14)
plt.xticks(rotation=45)
plt.legend(title='Cover Type')
plt.tight_layout()
plt.show()

# Plot 2: Compare Calculated Premium per Term across PostalCodes
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_cleaned, x='PostalCode', y='CalculatedPremiumPerTerm', palette='Set2')
plt.title('Comparison of Premium per Term Across Postal Codes', size=16)
plt.xlabel('Postal Code', size=14)
plt.ylabel('Calculated Premium per Term', size=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 3: Compare AutoMake across PostalCodes
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='PostalCode', hue='Make', palette='Set3')
plt.title('Distribution of Auto Make Across Postal Codes', size=16)
plt.xlabel('Postal Code', size=14)
plt.ylabel('Count', size=14)
plt.xticks(rotation=45)
plt.legend(title='Auto Make', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
